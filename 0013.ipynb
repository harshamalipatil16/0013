```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/college_student_management_data.csv')
print(df.major.unique())
print(df.major.value_counts())
color = {'Computer Science':'green', 'Arts':'blue', 'Engineering':'grey', 'Business':'red'}
plt.bar(df.major.unique(), df.major.value_counts(), color=color.values())
plt.xlabel('Major')
plt.ylabel('Number of Students')
plt.title('Number of Students in Each Major')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/netflix.csv')

df['Open'] = df['Open'].str.replace(',', '').astype(float)

print(df.Open.max())
print(df.Open.min())
print(df.Open.mean())
print(df.Open.median())
print(df.Open.std())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/penguins_size.csv')
df.species.value_counts().plot(kind='bar')
plt.xlabel('Species')
plt.ylabel('Number of Penguins')
plt.title('Number of Penguins in Each Species')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/penguins_size.csv')
df.island.value_counts().plot(kind='bar')
plt.xlabel('Island')
plt.ylabel('Number of Penguins')
plt.title('Number of Penguins in Each Island')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/penguins_size.csv')
plt.plot(df.culmen_length_mm, df.culmen_depth_mm, 'ro')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Culmen Depth (mm)')
plt.title('Relationship between Culmen Length and Depth')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/penguins_size.csv')
sns.histplot(df.culmen_length_mm, bins=20, kde=True)
plt.grid(True)
plt.show()
```
